```python
# Create DB, Table, Insert data and Fetch data USING SQLITE3
import sqlite3
```

```python
conn=sqlite3.connect('hrish.db')
```

```python
sql_create_tbl='''create table student
                (
                    student_id int primary key not null,
                    name text not null,
                    phn char(11) not null
                );'''
conn.execute(sql_create_tbl)
```

```python
sql_insert='''insert into student values(1,'Hrishagni','13136039672');'''
conn.execute(sql_insert)
conn.commit()
```

```python
select_sql='''select * from student;'''
cursor=conn.execute(select_sql)

for results in cursor:
    print(results)
```

```python
import requests,bs4
```

```python
# get facebook messenger app reviews from playstore in HTML format
url = "https://play.google.com/store/apps/details?id=com.facebook.orca&hl=en"
data=requests.get(url).text
soup=bs4.BeautifulSoup(data,"html.parser")
print(soup)
```
