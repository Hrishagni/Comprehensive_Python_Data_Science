```python
import pandas as pd
import numpy as np
```

```python
# Create a Pandas Series from a Numpy array
arr=np.array([2,3,4,5,6])
ser=pd.Series(arr)
print(arr)
print(ser)
print("Pandas Series has Index that starts with 0 unlike Numpy Arrays")
```

```python
# Create Pandas Series of type Datetime
dtser=pd.date_range(start='09/24/2021',end='10/5/2021')
print(dtser)
```

```python
# Slicing a Pandas Series
print("2nd element onwards:\n",ser[1:],"\n\n")
print("Last element:\t",ser[len(ser)-1:])
```

```python
# Changing index of a Pandas Series
new_ser=pd.Series(arr,index=['A','B','C','D','E'])
print(new_ser)
print("\nYou can create whatever index that you want to")
```
